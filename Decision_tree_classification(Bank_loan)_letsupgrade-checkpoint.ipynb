```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_excel(r'C:\Users\madhuri\Desktop\letsupgrade\Statistics_dinesh\Bank_Personal_Loan_Modelling.xlsx',sheet_name = 1)
```

```python
df.head()
```

```python
df = df.drop(['ID','ZIP Code'],axis = 1)
#print(df)
y = df['Personal Loan']
x = df.drop(['Personal Loan'],axis = 1)
print(x)
print(y)
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=8)
decision_tree.fit(x_train,y_train)
```

```python
y_pred = decision_tree.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
con = confusion_matrix(y_pred,y_test)
clas = classification_report(y_pred,y_test)
print(con)
print(clas)
```
