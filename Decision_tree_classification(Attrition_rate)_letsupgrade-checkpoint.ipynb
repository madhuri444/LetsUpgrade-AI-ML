```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv(r'C:\Users\madhuri\Desktop\letsupgrade\Day24\general_data.csv')
```

```python
df.head()
df = df.dropna()
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
objList = df.select_dtypes(include = "object").columns
for feat in objList:
    df[feat] = lb.fit_transform(df[feat].astype(str))

```

```python
df = df.drop(['EmployeeID','EmployeeCount','Over18'],axis = 1)
y = df['Attrition']
x = df.drop(['Attrition'],axis = 1)
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=8)
decision_tree.fit(x_train,y_train)
```

```python
y_pred = decision_tree.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
con = confusion_matrix(y_pred,y_test)
clas = classification_report(y_pred,y_test)
print(con)
print(clas)
```
